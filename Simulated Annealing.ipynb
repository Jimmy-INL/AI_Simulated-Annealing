```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import decimal
from mpl_toolkits import mplot3d
```

```python
def fungsiGraph(x1,X2):
    return -(abs(np.sin(x1)*np.cos(x2)*np.exp(abs(1-((np.sqrt(x1**2+x2**2))/np.pi)))))
def fungsi(x1,x2):
    return -(abs(math.sin(x1)*math.cos(x2)*math.exp(abs(1-((math.sqrt(x1**2+x2**2))/math.pi)))))
def RandomFunct(function):
    x1 = random.uniform(-10,10)
    x2 = random.uniform(-10,10)
    function.append(fungsi(x1,x2))
def getP(a,b):
    return decimal.Decimal(((-a)/b)).exp()
```

```python
CurrentState = []
RandomFunct(CurrentState)
BestOfFar = []
BestOfFar.append(CurrentState[-1])
T = 9999999999999999999999999999
Tend = 1
CoolRate = 0.999
```

```python
while T > Tend :
    newState = fungsi(random.uniform(-10,10),random.uniform(-10,10))
    deltaE = newState - CurrentState[-1] 
    if deltaE < 0 :
        CurrentState.append(newState)
        if newState < BestOfFar[-1]:
            BestOfFar.append(newState)
    else :
        r = random.uniform(0,1)
        p = getP(deltaE,T)
        if  p > r:
            CurrentState.append(newState)
    T = T * CoolRate
print("Hasil:", round(BestOfFar[-1],4))
```

```python
x1 = np.arange(-10.0, 10.0, 1)
x2 = np.arange(-10.0, 10.0, 1)
X1,X2 = np.meshgrid(x1,x2)
Fx = fungsiGraph(X1, X2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X1, X2, Fx, rstride=2, cstride=2,cmap='viridis', edgecolor='none')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('f(x)')
```
